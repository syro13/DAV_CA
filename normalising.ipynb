```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from itertools import combinations
```

```python
def minmax_normalise_dataset(path_to_csv, output_path):
    # Load the dataset
    df = pd.read_csv(path_to_csv)

    # Separate the country column (assumed to be the first)
    country_col = df.columns[0]
    country = df[[country_col]]
    data = df.drop(columns=[country_col])

    # Apply Min-Max scaling
    scaler = MinMaxScaler()
    data_scaled = scaler.fit_transform(data)

    # Reconstruct the dataframe
    df_scaled = pd.DataFrame(data_scaled, columns=data.columns)
    df_scaled.insert(0, country_col, country)

    # Save the normalised data
    df_scaled.to_csv(output_path, index=False)
    print(f"Min-Max normalised file saved to: {output_path}")

def plot_all_scatter_plots(csv_path):
    # Load the dataset
    df = pd.read_csv(csv_path)
    
    # Remove the first column (assumed to be Country or label column)
    numeric_df = df.select_dtypes(include='number')
    
    # Generate all pairwise scatter plots
    pairs = combinations(numeric_df.columns, 2)
    
    for x_var, y_var in pairs:
        plt.figure(figsize=(6, 4))
        plt.scatter(numeric_df[x_var], numeric_df[y_var], alpha=0.6)
        plt.xlabel(x_var)
        plt.ylabel(y_var)
        plt.title(f"{x_var} vs {y_var}")
        plt.grid(True)
        plt.tight_layout()
        plt.show()
```

```python
minmax_normalise_dataset("cleaned_sub_indices/access_connectivity.csv", "cleaned_sub_indices/access_connectivity_minmax.csv")
minmax_normalise_dataset("cleaned_sub_indices/digital_governance_participation.csv", "cleaned_sub_indices/digital_governance_participation_minmax.csv")
minmax_normalise_dataset("cleaned_sub_indices/digital_trade_economy.csv", "cleaned_sub_indices/digital_trade_economy_minmax.csv")
minmax_normalise_dataset("cleaned_sub_indices/education_human_capital.csv", "cleaned_sub_indices/education_human_capital_minmax.csv")
minmax_normalise_dataset("cleaned_sub_indices/infrastructure_security.csv", "cleaned_sub_indices/infrastructure_security_minmax.csv")
minmax_normalise_dataset("cleaned_sub_indices/innovation_rd.csv", "cleaned_sub_indices/innovation_rd_minmax.csv")
```

```python
plot_all_scatter_plots("cleaned_sub_indices/access_connectivity_minmax.csv")
plot_all_scatter_plots("cleaned_sub_indices/digital_governance_participation_minmax.csv")
plot_all_scatter_plots("cleaned_sub_indices/digital_trade_economy_minmax.csv")
plot_all_scatter_plots("cleaned_sub_indices/education_human_capital_minmax.csv")
plot_all_scatter_plots("cleaned_sub_indices/infrastructure_security_minmax.csv")
plot_all_scatter_plots("cleaned_sub_indices/innovation_rd_minmax.csv")
```
